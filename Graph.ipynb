```python
#  read quality scores
def readCsv(filename):
    file = open(filename, 'r')
    fileArr = []
    for line in file:
        fileArr.append(line.strip())
    return fileArr
```

```python
def readLexExperimentsScores():
    outputFilePath = './resources/OutputFiles/'
    file = 'qualityScores'
    fileArr = []
    for i in range(1,4):
        fileArr.append(file + '_' + str(i) + '.txt')
    threshTolValuesArr = []
    scoresArr = []
    for file in fileArr:
        threshTolValues = []
        scoreValues = []
        filename = outputFilePath + file
        fileContentArr = readCsv(filename)
        for line in fileContentArr:
            lineArr = line.split(',')
            threshTolVal = lineArr[0][4:-5]
            score = lineArr[1]
            threshTolValues.append(threshTolVal)
            scoreValues.append(score)
        scoresArr.append(scoreValues)
        threshTolValuesArr.append(threshTolValues)
    return scoresArr, threshTolValuesArr
```

```python
scoresArr, threshTolValuesArr = readLexExperimentsScores()
```

```python
#  convert array to dataframe
import pandas as pd
data = [
#     ('ThresholdToleranceCount', threshTolValuesArr[0] ),
       ('file1', scoresArr[0]),
       ('file2', scoresArr[1]),
       ('file3', scoresArr[2])]
#     ]
df = pd.DataFrame.from_items(data)
```

```python
def readQualityScores(outputFilePath):
#     outputFilePath = './resources/OutputFiles/'
    file = 'qualityScores.txt'
    lexScoreValues = []
    textScoreValues = []
    filename = outputFilePath + file
    fileContentArr = readCsv(filename)
    for line in fileContentArr:
        lineArr = line.split(',')
        lexScore = float(lineArr[1])
        textScore = float(lineArr[2])
        lexScoreValues.append(lexScore)
        textScoreValues.append(textScore)
    return lexScoreValues, textScoreValues
```

```python
#  convert array to dataframe
import pandas as pd
def getDataFrame(lexScoreValues, textScoreValues):
    data = [
           ('Lex Scores', lexScoreValues),
           ('Text Scores', textScoreValues)]

    df = pd.DataFrame.from_items(data)
    return df
```

```python
import matplotlib.pyplot as plt
# import matplotlib.axes.Axes as ax 
plt.style.use('fivethirtyeight')
def drawGraph(df, title, xticksArr, xticksTextArr):
    plt.figure(figsize=(30, 10))
    plt.title(title)
    plt.xticks([])
    plt.xticks(xticksArr, xticksTextArr)
    plt.xlabel('Files')
    plt.ylabel('Summarisation Quality Score')
    plt1 = plt.plot(df)
    plt.legend(plt1, df.columns)  
    plt.show()
```

```python
lexScores, textScores = readQualityScores('./resources/OutputFiles_CNN_994_2filter/')
scoresDF = getDataFrame(lexScores, textScores)
xticksArr = range(0,len(scoresDF)+10,100)
xticksTextArr = [str(i) for i in xticksArr]
drawGraph(scoresDF, 'Quality Scores comparison with 2 filters', xticksArr, xticksTextArr)
```
